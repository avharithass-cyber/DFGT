```python
import os
from PIL import Image
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, random_split, Dataset
from torchvision import transforms, models

# MultiModalTongueDataset class definition
class MultiModalTongueDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.image_paths = []
        self.labels = []

        for label, folder in enumerate(["Male_images", "Female_images"]):
            folder_path = os.path.join(root_dir, folder)
            for img in os.listdir(folder_path):
                self.image_paths.append(os.path.join(folder_path, img))
                self.labels.append(label)

        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img = Image.open(self.image_paths[idx]).convert("RGB")

        # Simulated modalities
        rgb = img
        # Modified nir and depth to be 3-channel (RGB) for ResNet compatibility
        nir = img.convert("RGB") # Changed from "L" to "RGB"
        depth = img.convert("RGB") # Changed from "L" to "RGB"

        if self.transform:
            rgb = self.transform(rgb)
            nir = self.transform(nir)
            depth = self.transform(depth)

        label = self.labels[idx]

        return rgb, nir, depth, label

# AttentionBlock and DFGTModel classes definition
# Attention Block
class AttentionBlock(nn.Module):
    def __init__(self, in_features):
        super().__init__()
        self.attn = nn.Sequential(
            nn.Linear(in_features, in_features),
            nn.ReLU(),
            nn.Sigmoid()
        )

    def forward(self, x):
        weights = self.attn(x)
        return x * weights


# Feature Extractor
def get_backbone():
    # Updated to use 'weights' argument instead of 'pretrained'
    model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)
    model.fc = nn.Identity()
    return model


class DFGTModel(nn.Module):
    def __init__(self, num_classes=2):
        super().__init__()

        self.rgb_net = get_backbone()
        self.nir_net = get_backbone()
        self.depth_net = get_backbone()

        self.attn_rgb = AttentionBlock(512)
        self.attn_nir = AttentionBlock(512)
        self.attn_depth = AttentionBlock(512)

        # Gated Fusion
        self.gate = nn.Sequential(
            nn.Linear(512*3, 512),
            nn.ReLU(),
            nn.Sigmoid()
        )

        # Heads
        self.id_head = nn.Linear(512, num_classes)
        self.spoof_head = nn.Linear(512, 2)

    def forward(self, rgb, nir, depth):
        f1 = self.attn_rgb(self.rgb_net(rgb))
        f2 = self.attn_nir(self.nir_net(nir))
        # Fixed typo: changed self.nir_net(depth) to self.depth_net(depth)
        f3 = self.attn_depth(self.depth_net(depth))

        fused = torch.cat([f1, f2, f3], dim=1)

        gate_weights = self.gate(fused)
        fused = fused[:, :512] * gate_weights  # gated fusion

        id_out = self.id_head(fused)
        spoof_out = self.spoof_head(fused)

        return id_out, spoof_out


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Transform
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

# Dataset
dataset = MultiModalTongueDataset(
    root_dir="/content/drive/MyDrive/DFGT Datasets Final ",
    transform=transform
)

# Split
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_data, val_data = random_split(dataset, [train_size, val_size])

train_loader = DataLoader(train_data, batch_size=16, shuffle=True)
val_loader = DataLoader(val_data, batch_size=16)

# Model
model = DFGTModel().to(device)

# Loss
criterion_id = nn.CrossEntropyLoss()
criterion_spoof = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

# Training Loop
epochs = 1

for epoch in range(epochs):
    model.train()
    total_loss = 0

    for rgb, nir, depth, labels in train_loader:
        rgb, nir, depth, labels = rgb.to(device), nir.to(device), depth.to(device), labels.to(device)

        optimizer.zero_grad()

        id_out, spoof_out = model(rgb, nir, depth)

        loss1 = criterion_id(id_out, labels)
        loss2 = criterion_spoof(spoof_out, labels)

        loss = loss1 + 0.5 * loss2
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {total_loss:.4f}")

    # Validation
    model.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for rgb, nir, depth, labels in val_loader:
            rgb, nir, depth, labels = rgb.to(device), nir.to(device), depth.to(device), labels.to(device)

            id_out, _ = model(rgb, nir, depth)
            _, predicted = torch.max(id_out, 1)

            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    acc = 100 * correct / total
    print(f"Validation Accuracy: {acc:.2f}%")

# Save Model
torch.save(model.state_dict(), "dfgt_model.pth")
```

```python
import os
import copy
import itertools
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
from PIL import Image, ImageFilter
import cv2
from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, f1_score,
                             precision_score, recall_score,
                             confusion_matrix)
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# ─────────────────────────
# 0. CONFIG
# ─────────────────────────
CFG = dict(
    # Paths – update to your local download of the Drive folder
    data_root        = "/tmp/tongue_dataset",          # root containing Male_images/ Female_images/
    img_size         = 224,
    batch_size       = 4,                            # small because dataset is tiny; scale up for full data
    num_epochs       = 30,
    lr               = 1e-4,
    weight_decay     = 1e-4,
    feature_dim      = 256,                          # embedding dim per branch
    num_classes      = 2,                            # Male / Female (replace with subject IDs for real data)
    spoof_classes    = 2,                            # real / spoof
    device           = "cuda" if torch.cuda.is_available() else "cpu",
    seed             = 42,
    dropout          = 0.3,
    temporal_frames  = 4                            # synthetic temporal sequence length
)

torch.manual_seed(CFG["seed"])
np.random.seed(CFG["seed"])

# Create dummy data for demonstration purposes
dummy_data_root = CFG["data_root"]
for gender in ["Male", "Female"]:
    os.makedirs(os.path.join(dummy_data_root, f"{gender}_images"), exist_ok=True)
    for i in range(15): # Create 15 dummy images for each gender (increased from 5)
        dummy_img_path = os.path.join(dummy_data_root, f"{gender}_images", f"dummy_img_{i:03d}.png")
        # Create a blank black image with a unique color in the center
        img = Image.new('RGB', (CFG["img_size"], CFG["img_size"]), color = 'black')
        pixels = img.load()
        # Add a unique color square to identify gender
        if gender == "Male":
            color = (255, 0, 0) # Red for Male
        else:
            color = (0, 0, 255) # Blue for Female

        # Draw a 20x20 square in the center
        for x_offset in range(-10, 10):
            for y_offset in range(-10, 10):
                x = CFG["img_size"] // 2 + x_offset
                y = CFG["img_size"] // 2 + y_offset
                if 0 <= x < CFG["img_size"] and 0 <= y < CFG["img_size"]:
                    pixels[x, y] = color

        img.save(dummy_img_path)
print(f"Dummy dataset created at {dummy_data_root}")


# ─────────────────────────
# 1. FEATURE EXTRACTION FROM RGB TONGUE IMAGES
#    Divides each image into 4 feature modalities
# ─────────────────────────

def extract_f1_appearance(img_np: np.ndarray) -> np.ndarray:
    """F1 – Appearance (RGB): colour texture, fissure patterns."""
    img = cv2.resize(img_np, (CFG["img_size"], CFG["img_size"]))
    img = img.astype(np.float32) / 255.0
    return img  # H×W×3


def extract_f2_vascular(img_np: np.ndarray) -> np.ndarray:
    """F2 – Vascular (NIR-simulated):
       Isolates sub-surface vascular patterns by enhancing the
       red channel (most sensitive to haemoglobin) with CLAHE.
    """
    img = cv2.resize(img_np, (CFG["img_size"], CFG["img_size"]))
    r_channel = img[:, :, 2]  # OpenCV BGR → index 2 is red

    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
    r_enhanced = clahe.apply(r_channel)

    # Build a 3-channel "NIR-like" image
    nir = np.stack([r_enhanced, r_enhanced, r_enhanced], axis=-1)
    nir = nir.astype(np.float32) / 255.0
    return nir  # H×W×3


def extract_f3_geometric(img_np: np.ndarray) -> np.ndarray:
    """F3 – Geometric / Depth-simulated:
       Captures tongue shape, boundary, and surface geometry via
       Canny edges + Gaussian-blurred depth approximation.
    """
    img = cv2.resize(img_np, (CFG["img_size"], CFG["img_size"]))
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Canny edge map
    edges = cv2.Canny(gray, 50, 150).astype(np.float32) / 255.0

    # Pseudo-depth: inverse of Laplacian sharpness (blurred → near, sharp → far)
    lap = cv2.Laplacian(gray, cv2.CV_32F)
    depth_proxy = cv2.GaussianBlur(np.abs(lap), (15, 15), 0)
    depth_proxy = depth_proxy / (depth_proxy.max() + 1e-8)

    # Morphological shape channel
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (7, 7))
    morph = cv2.morphologyEx(gray, cv2.MORPH_GRADIENT, kernel).astype(np.float32) / 255.0

    geo = np.stack([edges, depth_proxy, morph], axis=-1)
    return geo  # H×W×3


def extract_f4_temporal(img_np: np.ndarray,
                         n_frames: int = CFG["temporal_frames"]) -> np.ndarray:
    """F4 – Temporal (motion-simulated):
       Generates a synthetic temporal sequence by applying small
       affine jitter + brightness perturbations to simulate tongue
       micro-motion across frames. Returns mean-pooled frame.
    """
    img = cv2.resize(img_np, (CFG["img_size"], CFG["img_size"]))
    h, w = img.shape[:2]
    frames = []
    for i in range(n_frames):
        angle  = np.random.uniform(-3, 3)
        tx, ty = np.random.uniform(-5, 5), np.random.uniform(-5, 5)
        M = cv2.getRotationMatrix2D((w / 2, h / 2), angle, 1.0)
        M[0, 2] += tx; M[1, 2] += ty
        # Apply affine transformation to create 'warped'
        warped = cv2.warpAffine(img, M, (w, h))
        # Define 'alpha' for brightness perturbation
        alpha = np.random.uniform(0.8, 1.2)
        warped = np.clip(warped.astype(np.float32) * alpha, 0, 255).astype(np.uint8)
        frames.append(warped)
    temporal = np.mean(frames, axis=0).astype(np.float32) / 255.0
    return temporal  # H×W×3


FEATURE_EXTRACTORS = {
    "F1": extract_f1_appearance,
    "F2": extract_f2_vascular,
    "F3": extract_f3_geometric,
    "F4": extract_f4_temporal,
}


# ─────────────
# 2. DATASET
# ─────────────

class TongueDataset(Dataset):
    """
    Loads tongue images from:
        <data_root>/Male_images/*.png   → label 0
        <data_root>/Female_images/*.png → label 1

    For each image, all four feature maps are extracted on-the-fly
    so the dataset doesn't need pre-split feature folders.
    """
    def __init__(self, file_list, labels, augment=False):
        self.files   = file_list
        self.labels  = labels
        self.augment = augment
        self.to_tensor = transforms.ToTensor()   # H×W×C → C×H×W, float [0,1]

    def __len__(self):
        return len(self.files)

    def _load_bgr(self, path):
        img = cv2.imread(path)
        if img is None:
            # Fallback: create synthetic noise image for dummy/corrupt files
            img = np.random.randint(100, 200,
                                    (CFG["img_size"], CFG["img_size"], 3),
                                    dtype=np.uint8)
        if self.augment:
            img = self._augment(img)
        return img

    def _augment(self, img):
        """Basic augmentation: flip, rotation, colour jitter."""
        if np.random.rand() > 0.5:
            img = cv2.flip(img, 1)
        angle = np.random.uniform(-15, 15)
        h, w  = img.shape[:2]
        M = cv2.getRotationMatrix2D((w/2, h/2), angle, 1.0)
        img = cv2.warpAffine(img, M, (w, h))
        alpha = np.random.uniform(0.8, 1.2)
        img = np.clip(img.astype(np.float32) * alpha, 0, 255).astype(np.uint8)
        return img

    def __getitem__(self, idx):
        img_bgr = self._load_bgr(self.files[idx])
        features = {}
        for name, extractor in FEATURE_EXTRACTORS.items():
            feat = extractor(img_bgr)               # H×W×3, float32 [0,1]
            feat = torch.from_numpy(feat).permute(2, 0, 1)  # 3×H×W
            features[name] = feat
        label = torch.tensor(self.labels[idx], dtype=torch.long)
        # Spoof label: for real data use actual spoof annotations;
        # here all samples are "real" (label=0 means genuine)
        spoof_label = torch.tensor(0, dtype=torch.long)
        return features, label, spoof_label


def build_dataset(data_root: str):
    """
    Scans Male_images/ and Female_images/ folders and returns
    train/val/test splits with stratification.
    """
    class_dirs = {
        0: os.path.join(data_root, "Male_images"),
        1: os.path.join(data_root, "Female_images"),
    }
    files, labels = [], []
    for label, folder in class_dirs.items():
        if not os.path.isdir(folder):
            print(f"[WARN] Folder not found: {folder}")
            continue
        for fname in sorted(os.listdir(folder)):
            if fname.lower().endswith((".png", ".jpg", ".jpeg")):
                files.append(os.path.join(folder, fname))
                labels.append(label)

    if len(files) == 0:
        raise FileNotFoundError(
            f"No images found under {data_root}. "
            "Download the Drive folder and set CFG['data_root']."
        )

    # Stratified split: 70% train, 15% val, 15% test
    X_tr, X_tmp, y_tr, y_tmp = train_test_split(
        files, labels, test_size=0.30,
        stratify=labels, random_state=CFG["seed"]
    )
    X_val, X_te, y_val, y_te = train_test_split(
        X_tmp, y_tmp, test_size=0.50,
        stratify=y_tmp, random_state=CFG["seed"]
    )

    train_ds = TongueDataset(X_tr,  y_tr,  augment=True)
    val_ds   = TongueDataset(X_val, y_val, augment=False)
    test_ds  = TongueDataset(X_te,  y_te,  augment=False)

    print(f"Dataset  →  Train: {len(train_ds)} | Val: {len(val_ds)} | Test: {len(test_ds)}")
    return train_ds, val_ds, test_ds


# ───────────────────
# 3. MODEL COMPONENTS
# ───────────────────

class ChannelAttention(nn.Module):
    """Squeeze-and-Excitation channel attention."""
    def __init__(self, in_ch, reduction=16):
        super().__init__()
        self.avg_pool = nn.AdaptiveAvgPool2d(1)
        self.fc = nn.Sequential(
            nn.Linear(in_ch, max(in_ch // reduction, 4)),
            nn.ReLU(inplace=True),
            nn.Linear(max(in_ch // reduction, 4), in_ch),
            nn.Sigmoid(),
        )

    def forward(self, x):
        b, c, _, _ = x.shape
        w = self.avg_pool(x).view(b, c)
        w = self.fc(w).view(b, c, 1, 1)
        return x * w


class CNNBranch(nn.Module):
    """
    Single-modality CNN branch with attention.
    Uses MobileNetV2 as lightweight backbone + channel attention.
    """
    def __init__(self, feature_dim: int = CFG["feature_dim"]):
        super().__init__()
        backbone = models.mobilenet_v2(weights=None)
        backbone.features[0][0] = nn.Conv2d(
            3, 32, kernel_size=3, stride=2, padding=1, bias=False
        )
        self.backbone = backbone.features          # output: B×1280×7×7 for 224×224 input
        self.attention = ChannelAttention(1280)
        self.pool      = nn.AdaptiveAvgPool2d(1)
        self.proj      = nn.Sequential(
            nn.Flatten(),
            nn.Linear(1280, feature_dim),
            nn.BatchNorm1d(feature_dim),
            nn.ReLU(inplace=True),
            nn.Dropout(CFG["dropout"]),
        )

    def forward(self, x):
        feat = self.backbone(x)                    # B×1280×7×7
        feat = self.attention(feat)
        feat = self.pool(feat)                     # B×1280×1×1
        return self.proj(feat)                     # B×feature_dim


class GatedFeatureFusion(nn.Module):
    """
    Adaptive gated fusion:
        fused = Σ_i  gate_i ⊗ f_i
    where gate_i = softmax(W [f1∥f2∥…∥fn])
    """
    def __init__(self, n_modalities: int, feature_dim: int):
        super().__init__()
        self.n = n_modalities
        self.gate_net = nn.Sequential(
            nn.Linear(n_modalities * feature_dim, n_modalities),
            nn.Softmax(dim=-1),
        )

    def forward(self, feats: list):
        # feats: list of B×D tensors, len = n_modalities
        concat = torch.cat(feats, dim=-1)          # B×(n*D)
        gates  = self.gate_net(concat)             # B×n
        fused  = sum(gates[:, i:i+1] * feats[i]
                     for i in range(self.n))       # B×D
        return fused


class DFGTModel(nn.Module):
    """
    Full DFGT model with:
    - N CNN branches (one per active modality)
    - Gated feature fusion
    - Dual heads: ID classification + Spoof detection
    """
    def __init__(self,
                 active_modalities: list,
                 num_id_classes: int = CFG["num_classes"],
                 num_spoof_classes: int = CFG["spoof_classes"],
                 feature_dim: int = CFG["feature_dim"]):
        super().__init__()
        self.modalities = active_modalities
        self.n = len(active_modalities)
        assert self.n >= 1

        # One branch per active modality
        self.branches = nn.ModuleDict({
            m: CNNBranch(feature_dim) for m in active_modalities
        })
        # Fusion (only if multiple modalities)
        if self.n > 1:
            self.fusion = GatedFeatureFusion(self.n, feature_dim)
        # Heads
        self.id_head = nn.Sequential(
            nn.Linear(feature_dim, 128),
            nn.ReLU(inplace=True),
            nn.Dropout(CFG["dropout"]),
            nn.Linear(128, num_id_classes),
        )
        self.spoof_head = nn.Sequential(
            nn.Linear(feature_dim, 64),
            nn.ReLU(inplace=True),
            nn.Linear(64, num_spoof_classes),
        )

    def forward(self, feature_dict: dict):
        feats = [self.branches[m](feature_dict[m]) for m in self.modalities]
        if self.n == 1:
            fused = feats[0]
        else:
            fused = self.fusion(feats)
        return self.id_head(fused), self.spoof_head(fused)


# ───────────────────────
# 4. TRAINING & EVALUATION
# ───────────────────────

def train_one_epoch(model, loader, optimizer, device):
    model.train()
    id_criterion    = nn.CrossEntropyLoss()
    spoof_criterion = nn.CrossEntropyLoss()
    total_loss = 0.0
    for batch in loader:
        feat_dict, id_labels, spoof_labels = batch
        feat_dict   = {k: v.to(device) for k, v in feat_dict.items()}
        id_labels   = id_labels.to(device)
        spoof_labels= spoof_labels.to(device)

        optimizer.zero_grad()
        id_out, spoof_out = model(feat_dict)
        loss = id_criterion(id_out, id_labels) + \
               0.5 * spoof_criterion(spoof_out, spoof_labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / max(len(loader), 1)


@torch.no_grad()
def evaluate(model, loader, device):
    model.eval()
    all_preds, all_labels = [], []
    for batch in loader:
        feat_dict, id_labels, _ = batch
        feat_dict = {k: v.to(device) for k, v in feat_dict.items()}
        id_out, _ = model(feat_dict)
        preds = id_out.argmax(dim=1).cpu().numpy()
        all_preds.extend(preds)
        all_labels.extend(id_labels.numpy())

    avg = "binary" if CFG["num_classes"] == 2 else "macro"
    acc  = accuracy_score(all_labels, all_preds)
    f1   = f1_score(all_labels, all_preds, average=avg, zero_division=0)
    prec = precision_score(all_labels, all_preds, average=avg, zero_division=0)
    rec  = recall_score(all_labels, all_preds, average=avg, zero_division=0)
    return dict(accuracy=acc, f1=f1, precision=prec, recall=rec)


def train_model(modalities: list,
                train_ds, val_ds,
                epochs: int = CFG["num_epochs"],
                verbose: bool = True):
    """Train a DFGT model for a given subset of modalities."""
    device = CFG["device"]

    train_loader = DataLoader(train_ds, batch_size=CFG["batch_size"],
                              shuffle=True,  drop_last=True)
    val_loader   = DataLoader(val_ds,   batch_size=CFG["batch_size"],
                              shuffle=False, drop_last=False)

    model     = DFGTModel(modalities).to(device)
    optimizer = torch.optim.AdamW(model.parameters(),
                                  lr=CFG["lr"],
                                  weight_decay=CFG["weight_decay"])
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs)

    best_val_acc  = 0.0
    best_state    = None

    for epoch in range(1, epochs + 1):
        loss = train_one_epoch(model, train_loader, optimizer, device)
        val_metrics = evaluate(model, val_loader, device)
        scheduler.step()

        if val_metrics["accuracy"] > best_val_acc:
            best_val_acc = val_metrics["accuracy"]
            best_state   = copy.deepcopy(model.state_dict())

        if verbose and epoch % 5 == 0:
            print(f"  Epoch {epoch:3d}/{epochs} | loss={loss:.4f} "
                  f"| val_acc={val_metrics['accuracy']:.4f}")

    model.load_state_dict(best_state)
    return model


# ──────────────────
# 5. ABLATION STUDY
#    Trains & tests every combination of F1–F4
# ──────────────────

ALL_FEATURES = ["F1", "F2", "F3", "F4"]

def run_ablation(train_ds, val_ds, test_ds, epochs=CFG["num_epochs"]):
    """Run all 15 combinations of the 4 modalities."""
    results = {}
    combos  = []
    for r in range(1, len(ALL_FEATURES) + 1):
        for combo in itertools.combinations(ALL_FEATURES, r):
            combos.append(list(combo))

    for combo in combos:
        key = "".join(combo)
        print(f"\n[ABLATION] Modalities: {combo}")
        model = train_model(combo, train_ds, val_ds, epochs=epochs, verbose=True)

        test_loader = DataLoader(test_ds, batch_size=CFG["batch_size"], shuffle=False)
        metrics     = evaluate(model, test_loader, CFG["device"])
        results[key] = metrics

        print(f"  → Test  acc={metrics['accuracy']:.4f}  "
              f"f1={metrics['f1']:.4f}  "
              f"precision={metrics['precision']:.4f}  "
              f"recall={metrics['recall']:.4f}")
    return results


# ─────────────────────
# 6. FULL DFGT ENSEMBLE
#    Best performing: F1+F2+F3+F4 with ensemble averaging
# ─────────────────────

class DFGTEnsemble(nn.Module):
    """
    Ensemble of two independently-trained F1F2F3F4 models;
    final prediction = averaged softmax probabilities.
    """
    def __init__(self, models_list):
        super().__init__()
        self.models = nn.ModuleList(models_list)

    @torch.no_grad()
    def forward(self, feature_dict):
        probs_sum = None
        for m in self.models:
            id_out, _ = m(feature_dict)
            p = F.softmax(id_out, dim=-1)
            probs_sum = p if probs_sum is None else probs_sum + p
        return probs_sum / len(self.models)


def train_ensemble(train_ds, val_ds, n=2, epochs=CFG["num_epochs"]):
    """Train n full-feature models and wrap in ensemble."""
    models_list = []
    for i in range(n):
        print(f"\n[ENSEMBLE] Training model {i+1}/{n} …")
        m = train_model(ALL_FEATURES, train_ds, val_ds, epochs=epochs)
        models_list.append(m)
    return DFGTEnsemble(models_list)


@torch.no_grad()
def evaluate_ensemble(ensemble, loader, device):
    ensemble.eval()
    all_preds, all_labels = [], []
    for batch in loader:
        feat_dict, id_labels, _ = batch
        feat_dict = {k: v.to(device) for k, v in feat_dict.items()}
        probs = ensemble(feat_dict)
        preds = probs.argmax(dim=1).cpu().numpy()
        all_preds.extend(preds)
        all_labels.extend(id_labels.numpy())

    avg  = "binary" if CFG["num_classes"] == 2 else "macro"
    acc  = accuracy_score(all_labels, all_preds)
    f1   = f1_score(all_labels, all_preds, average=avg, zero_division=0)
    prec = precision_score(all_labels, all_preds, average=avg, zero_division=0)
    rec  = recall_score(all_labels, all_preds, average=avg, zero_division=0)
    return dict(accuracy=acc, f1=f1, precision=prec, recall=rec)


# ───────────────────
# 7. VISUALISE RESULTS
# ───────────────────

def plot_ablation(results: dict, save_path="ablation_results.png"):
    combos = list(results.keys())
    accs   = [results[k]["accuracy"] * 100 for k in combos]
    f1s    = [results[k]["f1"]        * 100 for k in combos]

    x = np.arange(len(combos))
    fig, ax = plt.subplots(figsize=(14, 5))
    w = 0.35
    ax.bar(x - w/2, accs, w, label="Accuracy (%)", color="#4C72B0")
    ax.bar(x + w/2, f1s,  w, label="F1-Score (%)", color="#DD8452")
    ax.set_xticks(x)
    ax.set_xticklabels(combos, rotation=45, ha="right", fontsize=9)
    ax.set_ylabel("Score (%)")
    ax.set_title("DFGT Ablation Study – Feature Combination Performance")
    ax.legend()
    ax.set_ylim(0, 110)
    for i, (a, f) in enumerate(zip(accs, f1s)):
        ax.text(i - w/2, a + 1, f"{a:.1f}", ha="center", fontsize=7)
        ax.text(i + w/2, f + 1, f"{f:.1f}", ha="center", fontsize=7)
    plt.tight_layout()
    plt.savefig(save_path, dpi=150)
    plt.close()
    print(f"[PLOT] Ablation chart saved → {save_path}")


def print_results_table(results: dict):
    print("\n" + "="*72)
    print(f"{'Combo':<14} {'Accuracy':>10} {'F1':>10} {'Precision':>12} {'Recall':>10}")
    print("-"*72)
    for k, v in sorted(results.items(), key=lambda x: x[1]["accuracy"]):
        print(f"{k:<14} {v['accuracy']*100:>9.2f}%  "
              f"{v['f1']*100:>9.2f}%  "
              f"{v['precision']*100:>11.2f}%  "
              f"{v['recall']*100:>9.2f}%")
    print("="*72)


# ───────────────────
# 8. MAIN PIPELINE
# ───────────────────

def main():
    print("=" * 60)
    print("  DFGT – Deep Fusion Framework for Tongue Biometrics")
    print("  University of North Texas")
    print("=" * 60)
    print(f"\nDevice : {CFG['device']}")
    print(f"Data   : {CFG['data_root']}\n")

    # ── Build dataset
    train_ds, val_ds, test_ds = build_dataset(CFG["data_root"])

    # ── Run ablation (all 15 feature subsets)
    print("\n[STEP 1] Ablation study across all feature combinations …")
    ablation_results = run_ablation(train_ds, val_ds, test_ds)
    print_results_table(ablation_results)
    plot_ablation(ablation_results)

    # ── Train final ensemble (F1+F2+F3+F4)
    print("\n[STEP 2] Training final ensemble model (F1+F2+F3+F4) …")
    ensemble = train_ensemble(train_ds, val_ds, n=2)

    test_loader = DataLoader(test_ds, batch_size=CFG["batch_size"], shuffle=False)
    final_metrics = evaluate_ensemble(ensemble, test_loader, CFG["device"])

    print("\n── Final Ensemble Test Results ──────────────────────")
    for k, v in final_metrics.items():
        print(f"  {k:<12}: {v*100:.2f}%")
    print("───────────────────────────────────")

    # ── Save ensemble
    torch.save(ensemble.state_dict(), "dfgt_ensemble.pth")
    print("\n[SAVED] Model weights → dfgt_ensemble.pth")


if __name__ == "__main__":
    main()
```
